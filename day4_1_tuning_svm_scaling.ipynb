```python
# 하이퍼 파라미터 튜닝 실습
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train ,X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify = cancer.target, random_state=66)
```

```python
X_train.shape, X_test.shape
```

```python
X_train[0]
```

```python
np.set_printoptions(precision=3,suppress=True)
```

```python
X_train[0]
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# 하이퍼 파라미터 튜닝을 반복문으로 함

train_acc = [] # 훈련 정확도
test_acc = [] # 테스트 정확도

n= range(1,11)
for i in n:
    clf = KNeighborsClassifier(n_neighbors=i) # i를 증가시키면서 훈련 데이터 테스트 데이터 정확도 분석
    clf.fit(X_train, y_train) 
    train_acc.append(clf.score(X_train, y_train))
    test_acc.append(clf.score(X_test, y_test))
```

```python
train_acc
```

```python
test_acc
```

```python
# train_acc, test_acc 시각화

plt.plot(n, train_acc, label='train')
plt.plot(n, test_acc,label = 'test')
plt.legend()
plt.show()
# test 점수가 가자 높은게 적당하다
```

```python
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train,y_train)
print("훈련 점수 : ", knn.score(X_train, y_train))
print("테스트 점수 : ", knn.score(X_test, y_test))
```
